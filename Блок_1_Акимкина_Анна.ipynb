###Работа с input() и числами
Когда вы запрашиваете данные у пользователя с помощью функции input(), они всегда возвращаются в виде строк. Это может вызвать проблемы при попытке выполнить математические операции, так как Python будет трактовать числа как текст.

###Неправильный пример: сложение строк вместо чисел

````python
number3 = 15
number4 = 20

sum = number3 + number4
print(sum)
````

###Правильный пример: приведение к числовому типу
Чтобы сложить числа правильно, нужно преобразовать строки в числа с помощью функции int().

````python
a = int(input())
b = int(input())

print(a + b)
````

````python
number1 = int(input("Введите первое число: "))
number2 = int(input("Введите второе число: "))

#измените код так чтобы складывалось три числа

result = number1 + number2
print("Результат сложения:", result)  # Корректный результат
````

### Задание 2:
1. Напишите программу, которая запрашивает у пользователя его имя и год рождения, а затем выводит сообщение: "Привет, [имя], тебе [возраст] лет". Для расчета возраста вычтите год рождения из текущего года.
2. Модифицируйте программу так, чтобы она предлагала пользователю ввести текущий год (это поможет закрепить работу с несколькими входными данными).

````python

name = input("Как вас зовут? ")
birth_year = int(input("Введите ваш год рождения: "))
current_year = int(input("Введите текущий год: "))
age = current_year - birth_year

print(f"Привет, {name}, тебе {age} лет!")
````


### Задание 4:

````python

length = float(input("Введите длину прямоугольника: "))
width = float(input("Введите ширину прямоугольника: "))
area = length * width
perimeter = 2 * (length + width)
print(f"Площадь прямоугольника равна: {area}")
print(f"Периметр прямоугольника равен: {perimeter}")

````
